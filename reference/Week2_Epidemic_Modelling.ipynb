## **What Is Epidemic Modeling?**

Epidemic modeling is the use of mathematical and statistical models to understand and predict the spread of infectious diseases.

**Core Goals of Epidemic Models**

Epidemic models are used to:

***Understand transmission dynamics***
→ How does a disease spread through a population?

***Identify key drivers***
→ Which factors (contact rate, recovery rate, immunity, interventions) influence spread?

***Support public health decision-making***
→ Vaccination policies, screening programs, lockdowns, resource allocation

### **Epidemic models are decision-support tools, not just prediction tools.**

## **2. Historical Context (Why These Models Exist)**

Key milestones:

**Daniel Bernoulli (1760)**
→ Showed vaccination increased life expectancy (smallpox)

**William Farr (19th century)**
→ First to fit epidemic curves to data

**En'ko (1889)**
→ Introduced discrete stochastic models for measles

**Kermack & McKendrick (1927)**
→ Introduced the SIR differential equation model

### **Modern epidemic modeling is built on both biology and mathematics.**

## 4. **Mechanistic Models**

**What Are Mechanistic Models?**

Mechanistic models explicitly describe the biological mechanisms that drive disease transmission.

They answer:

- Why does the disease spread?

- How do individuals move between disease states?

They are ***hypothesis-driven***.

### 4.1 **Compartmental Modeling (Mechanistic)**
**Core Idea**

A population is divided into compartments, representing disease states.

Individuals move between compartments over time.

**4.1.1 The SIR Model**

The classical SIR model has three compartments:

| Compartment | Meaning                       |
| ----------- | ----------------------------- |
| S           | Susceptible                   |
| I           | Infected (infectious)         |
| R           | Removed (recovered or immune) |

***Assumption: Recovered individuals gain permanent immunity.***

**Why SIR Is Popular?**

Simple, Few parameters, Captures core epidemic behavior

**Why SIR Is Often Insufficient?**

No latent period, Assumes homogeneous mixing,Ignores behavioral and demographic heterogeneity

**4.1.2 SEIR Model (Latent Period)**

Some diseases have a latent (exposed) period: Infected but not yet infectious

The **SEIR model** adds an **E** compartment.
